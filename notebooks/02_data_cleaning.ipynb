# 🧹 Notebook 02: Data Cleaning
## Customer Churn Prediction
- Fix TotalCharges data type (Object -> Float)
- Handle missing values
- Remove duplicates
- Standardize column names

```python
# ════════════════════════════════════════
# CELL 2: Imports & Setup
# ════════════════════════════════════════
import pandas as pd
import numpy as np
import os
from pathlib import Path

print("✅ Libraries imported!")

# Define Paths (Auto-detect Kaggle vs VS Code)
if os.path.exists('/kaggle/working'):
    INPUT_PATH = '/kaggle/input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv'
    OUTPUT_PATH = Path('/kaggle/working/data/processed/cleaned_data.csv')
else:
    INPUT_PATH = Path('../data/raw/telco_churn.csv')
    OUTPUT_PATH = Path('../data/processed/cleaned_data.csv')

print(f"📂 Input: {INPUT_PATH}")
print(f"📂 Output: {OUTPUT_PATH}")
```

```python
# ════════════════════════════════════════
# CELL 3: Load Raw Data
# ════════════════════════════════════════
df = pd.read_csv(INPUT_PATH)
print(f"✅ Loaded dataset: {df.shape}")
print(f"🔍 Columns: {list(df.columns)}")
```

```python
# ════════════════════════════════════════
# CELL 4: Fix TotalCharges Data Type
# ════════════════════════════════════════
print("🔧 Fixing TotalCharges data type...")
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
nulls_before = df['TotalCharges'].isnull().sum()
print(f"⚠️ Missing values in TotalCharges after conversion: {nulls_before}")
```

```python
# ════════════════════════════════════════
# CELL 5: Handle Missing Values
# ════════════════════════════════════════
print("🔧 Handling missing values...")
median_val = df['TotalCharges'].median()
df['TotalCharges'].fillna(median_val, inplace=True)
print(f"✅ Missing values remaining: {df.isnull().sum().sum()}")
```

```python
# ════════════════════════════════════════
# CELL 6: Remove Duplicates
# ════════════════════════════════════════
print("🔧 Checking for duplicates...")
dups = df.duplicated().sum()
print(f"⚠️ Duplicate rows found: {dups}")
if dups > 0:
    df.drop_duplicates(inplace=True)
    print("✅ Duplicates removed.")
else:
    print("✅ No duplicates found.")
```

```python
# ════════════════════════════════════════
# CELL 7: Standardize Column Names
# ════════════════════════════════════════
df.columns = df.columns.str.lower().str.replace(' ', '_')
print(f"✅ Columns standardized: {list(df.columns)}")
```

```python
# ════════════════════════════════════════
# CELL 8: Final Verification
# ════════════════════════════════════════
print("="*50)
print("📊 CLEANED DATA SUMMARY")
print("="*50)
print(f"Shape: {df.shape}")
print(f"Data Types:\n{df.dtypes}")
print(f"Missing Values: {df.isnull().sum().sum()}")
print(f"Duplicates: {df.duplicated().sum()}")
```

```python
# ════════════════════════════════════════
# CELL 9: Save Cleaned Data
# ════════════════════════════════════════
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(OUTPUT_PATH, index=False)
print(f"💾 Saved cleaned data to: {OUTPUT_PATH}")
print("✅ Notebook 02 Complete!")
```
