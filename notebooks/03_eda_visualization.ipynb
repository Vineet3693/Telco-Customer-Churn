```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
print("✅ Libraries imported!")

if os.path.exists('/kaggle/working'):
    DATA_PATH = '/kaggle/working/data/processed/cleaned_data.csv'
    FIG_DIR = Path('/kaggle/working/reports/figures')
else:
    DATA_PATH = Path('../data/processed/cleaned_data.csv')
    FIG_DIR = Path('../reports/figures')

FIG_DIR.mkdir(parents=True, exist_ok=True)
df = pd.read_csv(DATA_PATH)
print(f"✅ Data loaded: {df.shape}")
```

```python
plt.figure(figsize=(8, 6))
churn_counts = df['churn'].value_counts()
bars = plt.bar(churn_counts.index, churn_counts.values, color=['#2ecc71', '#e74c3c'])
plt.title('Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Count')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height, f'{height}', ha='center', va='bottom')
plt.savefig(FIG_DIR / 'churn_distribution.png')
print("✅ Saved: churn_distribution.png")
plt.show()
```

```python
num_cols = ['tenure', 'monthlycharges', 'totalcharges']
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for i, col in enumerate(num_cols):
    sns.boxplot(x='churn', y=col, data=df, ax=axes[i], palette='Set2')
    axes[i].set_title(f'{col} vs Churn')
plt.tight_layout()
plt.savefig(FIG_DIR / 'numerical_vs_churn.png')
print("✅ Saved: numerical_vs_churn.png")
plt.show()
```

```python
df_corr = df.copy()
df_corr['churn_encoded'] = df_corr['churn'].map({'Yes': 1, 'No': 0})
plt.figure(figsize=(12, 10))
numeric_df = df_corr.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.savefig(FIG_DIR / 'correlation_heatmap.png')
print("✅ Saved: correlation_heatmap.png")
plt.show()
```
