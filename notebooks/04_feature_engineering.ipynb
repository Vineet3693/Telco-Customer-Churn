# 🛠️ Notebook 04: Feature Engineering
## Customer Churn Prediction
- Create tenure_group
- Create avg_monthly_to_total_ratio
- Create is_long_term_customer
- Create num_services
- Create has_no_support

```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
print("✅ Libraries imported!")

if os.path.exists('/kaggle/working'):
    INPUT_PATH = '/kaggle/working/data/processed/cleaned_data.csv'
    OUTPUT_PATH = Path('/kaggle/working/data/processed/featured_data.csv')
else:
    INPUT_PATH = Path('../data/processed/cleaned_data.csv')
    OUTPUT_PATH = Path('../data/processed/featured_data.csv')

df = pd.read_csv(INPUT_PATH)
print(f"✅ Loaded: {df.shape}")
```

```python
def get_tenure_group(tenure):
    if tenure <= 12: return '0-12'
    elif tenure <= 24: return '12-24'
    elif tenure <= 48: return '24-48'
    else: return '48-72'

df['tenure_group'] = df['tenure'].apply(get_tenure_group)
print("✅ Created: tenure_group")
print(df['tenure_group'].value_counts())
```

```python
df['avg_monthly_to_total_ratio'] = df.apply(
    lambda x: x['monthlycharges'] / x['totalcharges'] if x['totalcharges'] > 0 else 0, axis=1)
print("✅ Created: avg_monthly_to_total_ratio")
```

```python
df['is_long_term_customer'] = (df['tenure'] > 36).astype(int)
print("✅ Created: is_long_term_customer")
print(df['is_long_term_customer'].value_counts())
```

```python
service_cols = ['phoneservice', 'multiplelines', 'internetservice', 'onlinesecurity', 'onlinebackup', 'deviceprotection', 'techsupport', 'streamingtv', 'streamingmovies']
def count_services(row):
    count = 0
    for col in service_cols:
        val = str(row[col]).lower()
        if val not in ['no', 'no phone service', 'no internet service', 'nan']:
            count += 1
    return count
df['num_services'] = df.apply(count_services, axis=1)
print("✅ Created: num_services")
```

```python
df['has_no_support'] = ((df['onlinesecurity'] == 'No') & (df['techsupport'] == 'No')).astype(int)
print("✅ Created: has_no_support")
```

```python
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(OUTPUT_PATH, index=False)
print(f"💾 Saved featured data to: {OUTPUT_PATH}")
print(f"📊 Final Shape: {df.shape}")
print("✅ Notebook 04 Complete!")
```
