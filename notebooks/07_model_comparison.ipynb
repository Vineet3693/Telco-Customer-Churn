## Customer Churn Prediction
- Bar plots for metrics
- Save comparison figure

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path
sns.set_style("whitegrid")
print("✅ Libraries imported!")

if os.path.exists('/kaggle/working'):
    RESULTS_PATH = '/kaggle/working/models/model_results.csv'
    FIG_DIR = Path('/kaggle/working/reports/figures')
else:
    RESULTS_PATH = Path('../models/model_results.csv')
    FIG_DIR = Path('../reports/figures')

df_results = pd.read_csv(RESULTS_PATH)
print(f"✅ Loaded results: {df_results.shape}")
```

```python
metrics = ['Accuracy', 'Precision', 'Recall', 'F1', 'ROC-AUC']
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
axes = axes.flatten()
for i, metric in enumerate(metrics):
    sorted_df = df_results.sort_values(metric, ascending=False)
    sns.barplot(x=metric, y='Model', data=sorted_df, ax=axes[i], palette='viridis')
    axes[i].set_title(f'Model Comparison by {metric}')
plt.tight_layout()
plt.savefig(FIG_DIR / 'model_comparison.png')
print("✅ Saved: model_comparison.png")
plt.show()
```
